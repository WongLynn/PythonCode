```python
import pymongo
import charts
client = pymongo.MongoClient("localhost", 27017)
db = client.data_mining
```

```python
def BubbleSort(seq):
    length = len(seq)
    for i in range(0,length):
        for j in range(i+1,length):
            if seq[i] < seq[j]:
                seq[i],seq[j] = seq[j],seq[i]
    return seq
friends_num_list = []
for item in db.renren_data.find():
    friends_num_list.append(int(item['friends']))
top_ten = BubbleSort(friends_num_list)[:10]
series = []
for item in db.renren_data.find():
    for i in top_ten:
        if int(i) == int(item['friends']):
            data = {
                'data' : [item['friends']],
                'name' : item['name'],
                'level' : item['level'],
                'type': 'column'
            }
            series.append(data)
options = {
    'chart'   : {'zoomType':'xy'},
    'title'   : {'text': 'Top ten'},
    'subtitle': {'text': 'friends_number'},
    'yAxis'   : {'title': {'text': 'Friends_number'}}
}
charts.plot(series,options=options,show='inline')
```
